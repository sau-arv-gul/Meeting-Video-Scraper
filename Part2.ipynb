```python
from yt_dlp import YoutubeDL
from yt_dlp.utils import DownloadError, ExtractorError

def check_url_downloadable(url):
    """
    Check if a single URL is downloadable by yt-dlp.
    Returns True if downloadable, False otherwise.
    """
    ydl_opts = {
        'quiet': True,
        'skip_download': True,
        'noplaylist': True,
    }
    try:
        with YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=False)
            return 'url' in info or 'formats' in info
    except (DownloadError, ExtractorError, Exception):
        return False

def check_urls_downloadable(url_list):
    """
    For a list of URLs, returns a dictionary {url: True/False}
    indicating whether each URL is downloadable by yt-dlp.
    """
    result = {}
    for url in url_list:
        result[url] = check_url_downloadable(url)
    return result

```

```python


from seleniumwire import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import requests
import chromedriver_autoinstaller

def get_zoomgov_video_url_and_headers(zoom_url):
    chromedriver_autoinstaller.install()

    chrome_options = Options()
    chrome_options.add_argument("--headless=new")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    driver = webdriver.Chrome(options=chrome_options)

    try:
        driver.get(zoom_url)
        time.sleep(8)  # Wait for the video to load

        video_element = driver.find_element(By.TAG_NAME, "video")
        video_url = video_element.get_attribute("src")

        if not video_url:
            raise Exception("Video URL not found.")

        cookies = {cookie['name']: cookie['value'] for cookie in driver.get_cookies()}
        headers = {
            "Referer": zoom_url,
            "User-Agent": driver.execute_script("return navigator.userAgent;")
        }
        print("üéØ Video URL:", video_url)
        return video_url, {"headers": headers, "cookies": cookies}

    finally:
        driver.quit()

def download_video_with_session(url, session_data, filename):
    response = requests.get(url, headers=session_data["headers"], cookies=session_data["cookies"], stream=True)

    if response.status_code != 200:
        print("‚ùå Download failed. Status Code:", response.status_code)
        return False

    with open(filename, 'wb') as f:
        for chunk in response.iter_content(8192):
            f.write(chunk)

    print(f"‚úÖ Video saved as {filename}")
    return True



link = "https://www.zoomgov.com/rec/share/vCZnQM5bgMzY7_n4lbQXYnVqsvPj49Ce-R0hMjMFPyG4FUC1HbSyQAZ9uPpRKDvV._6ZMrf7BXZzx6_RK?startTime=1709655569000"


video_url, session = get_zoomgov_video_url_and_headers(link)

# Optional: check manually if it's authorized
download_video_with_session(video_url, session,"zoom_recording.mp4")
```

```python
import requests
import re

def Direct_URL(url):
  resp = requests.get(url)
  html = resp.text

  # Regex to find the downloadLinks array inside the script tag
  match = re.search(r'downloadLinks\s*=\s*(\[[^\]]+\])', html)

  direct_url = ""
  if match:
      # Extract the JSON-like array string
      links_str = match.group(1)

      # Extract URLs inside the array string
      urls = re.findall(r'"(https:\\/\\/archive-video\.granicus\.com[^"]+)"', links_str)

      # Unescape the URLs (remove backslashes)
      urls = [u.replace('\\/', '/') for u in urls]

      if urls:
          print("Direct video URL found:", urls[0])
          direct_url = urls[0]
      else:
          print("No URLs found inside downloadLinks.")
  else:
      print("downloadLinks not found in page source.")
  return direct_url

link = "https://snohomish.legistar.com/Video.aspx?Mode=Granicus&ID1=9188&Mode2=Video"
direct_url = Direct_URL(link)
download.append(direct_url)
```

```python
import requests
from bs4 import BeautifulSoup
import re
def extract_video_url(page_url: str) -> str | None:
    """
    Fetch the page at `page_url` and extract the video_url value from a <script> tag.

    Args:
        page_url (str): The URL of the webpage to scrape.

    Returns:
        str or None: The extracted video URL if found, else None.
    """
    resp = requests.get(page_url)
    if resp.status_code != 200:
        print(f"Failed to fetch page, status code: {resp.status_code}")
        return None

    soup = BeautifulSoup(resp.text, 'html.parser')

    for script in soup.find_all('script'):
        if script.string:
            match = re.search(r'video_url\s*=\s*"([^"]+)"', script.string)
            if match:
                return match.group(1)
    return None

# Example usage
link = "https://multnomah.granicus.com/MediaPlayer.php?view_id=3&clip_id=3097"
video_url = extract_video_url(link)
if video_url:
    print("Extracted video URL:", video_url)
    download.append(video_url)
else:
    print("Video URL not found.")
```

```python



from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import chromedriver_autoinstaller
import time

def get_direct_video_url(url):
    # Install ChromeDriver if not already installed
    chromedriver_autoinstaller.install()

    # Setup Chrome options
    options = Options()
    # Remove comment below to run headless
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")

    # Launch browser
    driver = webdriver.Chrome(options=options)
    driver.get(url)

    try:
        # Wait for the video player to be ready
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.ID, "EventDetailsVideo"))
        )
        time.sleep(2)


        # Wait for play button and click it
        play_btn = WebDriverWait(driver, 20).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, ".jw-icon-display"))
        )
        play_btn.click()

        # Wait for the video tag to appear with `src`
        video_elem = WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "video.jw-video"))
        )

        # Get the direct video URL
        video_url = video_elem.get_attribute("src")
        return video_url

    except Exception as e:
        print(f"Error: {e}")
        return None

    finally:
        driver.quit()


# Example usage
link = "https://traviscotx.portal.civicclerk.com/event/2583/media"
video_url = get_direct_video_url(link)
download.append(video_url)
print("Direct video URL:", video_url)
```
